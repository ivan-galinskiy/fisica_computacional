# Mandar parámetros a funciones

```julia
f(x,t) = -α*x
```

```julia
f(1., 0.)
```

```julia
α = 3.  # ¡NO ESTÁ BONITO! ¡NO LO UTILICES!
```

```julia
f(1., 0.)
```

En lugar de

euler(f, ...)

usamos:

```julia
f(x,t, α) = -α*x   # en matemáticas: f(x, t; α)
```

```julia
(x,t) -> f(x, t, 3.)  # función anónima
```

```julia
euler((x,t)->f(x,t,3.), ...)
```

```julia
g = (x,t) -> f(x, t, 3.)  # función anónima
euler(g, ...)
```
